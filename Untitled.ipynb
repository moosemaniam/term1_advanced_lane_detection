```python
enable_sanity_checks=False
plt_index=1
flag=1
left_fitx_last_good=None
right_fitx_last_good=None
ploty_last_good=None
left_curverad_last_good=None
right_curverad_last_good=None
offcenter_last_good=None
left_fit_last_good=None
right_fit_last_good=None
frame_no=0
enable_plot=1

def similarity(x,y):
    """ http://dataaspirant.com/2015/04/11/five-most-popular-similarity-measures-implementation-in-python/
    Euclidean distance"""
    return sqrt(sum(pow(a-b,2) for a, b in zip(x, y)))
    

def calculate_y(curve,x):
    """Calculate y pixel coordinate value, given curve and x pixel coordinate"""
    return ( curve[0]*x**2+curve[1]*x + curve[0])

def bad_curves_check(curve_latest,curve_last_good):
    """Checks if a given curve is too different from
    a last known good curve and returns true if so"""
    threshold=100.0
    val = similarity(curve_latest,curve_last_good)
    if(val > threshold):
        return True
    
    return False
    
    
    print("diff",a_diff,b_diff,c_diff)
    if(a_diff > threshold or b_diff > threshold or c_diff > threshold):
        print("Too much of a change")
        return True
    
    return False
    

def plot_image(img,string):
    global plt_index
    global enable_plot
    if(enable_plot!=1):
        return
    plt.subplot(11, 2, plt_index)
    plt.title(string)
    plt.imshow(img,cmap='gray')
    plt.axis('off')
    plt_index +=1
    
def windowed_lane_finder(img):
    global left_fitx_last_good
    global right_fitx_last_good
    global ploty_last_good
    global left_curverad_last_good
    global right_curverad_last_good
    global offcenter_last_good
    global left_fit_last_good
    global right_fit_last_good
    global flag
    global enable_sanity_checks
    
    binary_warped = img
    histogram = np.sum(binary_warped[binary_warped.shape[0]/2:,:], axis=0)
    out_img = np.dstack((binary_warped, binary_warped, binary_warped))*255
    midpoint = np.int(histogram.shape[0]/2)
    leftx_base = np.argmax(histogram[:midpoint])
    rightx_base = np.argmax(histogram[midpoint:]) + midpoint
    
    # Choose the number of sliding windows
    nwindows = 9
    # Set height of windows
    window_height = np.int(binary_warped.shape[0]/nwindows)
    # Identify the x and y positions of all nonzero pixels in the image
    nonzero = binary_warped.nonzero()
    nonzeroy = np.array(nonzero[0])
    nonzerox = np.array(nonzero[1])
    # Current positions to be updated for each window
    leftx_current = leftx_base
    rightx_current = rightx_base
    # Set the width of the windows +/- margin
    margin = 100
    # Set minimum number of pixels found to recenter window
    minpix = 50
    # Create empty lists to receive left and right lane pixel indices
    left_lane_inds = []
    right_lane_inds = []
    
    # Step through the windows one by one
    for window in range(nwindows):
        # Identify window boundaries in x and y (and right and left)
        win_y_low = binary_warped.shape[0] - (window+1)*window_height
        win_y_high = binary_warped.shape[0] - window*window_height
        win_xleft_low = leftx_current - margin
        win_xleft_high = leftx_current + margin
        win_xright_low = rightx_current - margin
        win_xright_high = rightx_current + margin
        # Draw the windows on the visualization image
        cv2.rectangle(out_img,(win_xleft_low,win_y_low),(win_xleft_high,win_y_high),(0,255,0), 2) 
        cv2.rectangle(out_img,(win_xright_low,win_y_low),(win_xright_high,win_y_high),(0,255,0), 2) 
        # Identify the nonzero pixels in x and y within the window
        good_left_inds = ((nonzeroy >= win_y_low) & (nonzeroy < win_y_high) & (nonzerox >= win_xleft_low) & (nonzerox < win_xleft_high)).nonzero()[0]
        good_right_inds = ((nonzeroy >= win_y_low) & (nonzeroy < win_y_high) & (nonzerox >= win_xright_low) & (nonzerox < win_xright_high)).nonzero()[0]
        # Append these indices to the lists
        left_lane_inds.append(good_left_inds)
        right_lane_inds.append(good_right_inds)
        # If you found > minpix pixels, recenter next window on their mean position
        if len(good_left_inds) > minpix:
            leftx_current = np.int(np.mean(nonzerox[good_left_inds]))
        if len(good_right_inds) > minpix:        
            rightx_current = np.int(np.mean(nonzerox[good_right_inds]))
        # Concatenate the arrays of indices
    left_lane_inds = np.concatenate(left_lane_inds)
    right_lane_inds = np.concatenate(right_lane_inds)

    
    # Extract left and right line pixel positions
    leftx = nonzerox[left_lane_inds]
    lefty = nonzeroy[left_lane_inds] 
    rightx = nonzerox[right_lane_inds]
    righty = nonzeroy[right_lane_inds] 

    #if the pixel positions come up empty, then just send back the 
    #last good line found
    if(enable_sanity_checks == True and (len(leftx) == 0 or len(lefty)==0 or len(rightx)==0 or len(righty)==0)):    
        return left_fitx_last_good,right_fitx_last_good,ploty_last_good,left_curverad_last_good,right_curverad_last_good,offcenter_last_good
    
    #If we found lines, continue
    # Fit a second order polynomial to each
    left_fit = np.polyfit(lefty, leftx, 2)
    right_fit = np.polyfit(righty, rightx, 2)
    
    #No sanity check initially
    if(flag==1):
        flag=0
        left_fit_last_good = copy.deepcopy(left_fit)
        right_fit_last_good = copy.deepcopy(right_fit)
        
    #Check for sanity
    if(enable_sanity_checks == True and (bad_curves_check(left_fit,left_fit_last_good) or bad_curves_check(right_fit,right_fit_last_good)) ):
        return left_fitx_last_good,right_fitx_last_good,ploty_last_good,left_curverad_last_good,right_curverad_last_good,offcenter_last_good
    ploty = np.linspace(0, binary_warped.shape[0]-1, binary_warped.shape[0] )
    left_fitx = left_fit[0]*ploty**2 + left_fit[1]*ploty + left_fit[2]
    right_fitx = right_fit[0]*ploty**2 + right_fit[1]*ploty + right_fit[2]

    
    #Figure out the curve
    # Define conversions in x and y from pixels space to meters
    ym_per_pix = 30/720 # meters per pixel in y dimension
    xm_per_pix = 3.7/700 # meters per pixel in x dimension

    y_eval = np.max(lefty)
    # Fit new polynomials to x,y in world space
   
    left_fit_cr = np.polyfit(lefty*ym_per_pix, leftx*xm_per_pix, 2)
    right_fit_cr = np.polyfit(righty*ym_per_pix, rightx*xm_per_pix, 2)
    # Calculate the new radii of curvature
    left_curverad = ((1 + (2*left_fit_cr[0]*y_eval*ym_per_pix + left_fit_cr[1])**2)**1.5) / np.absolute(2*left_fit_cr[0])
    right_curverad = ((1 + (2*right_fit_cr[0]*y_eval*ym_per_pix + right_fit_cr[1])**2)**1.5) / np.absolute(2*right_fit_cr[0])
    # Now our radius of curvature is in meters
    
    #Find the deviation of the car
    """1) Figure out x co-ordinates of laneright and lane left near the car
    2) Mid point of these x co-ordinates is the middle of the road
    3) xsize/2 of the image gives the x cordinates of the car
    4) Diff between values 2) and 3) gives the deviation of the car off center
    """
    left_index_x = np.argmax(lefty)
    left_x_coordinate = leftx[left_index_x]
    
    right_index_x = np.argmax(righty)
    right_x_coordinate = rightx[right_index_x]
    
    center_x_coordinate = (img.shape[1]/2)
    
    #From the review
    middle_of_lane_x_co_ordinate = (right_x_coordinate + left_x_coordinate)/2
    
    #Convert to meters
    offcenter = np.abs(xm_per_pix * (middle_of_lane_x_co_ordinate - center_x_coordinate))
    
    left_fit_last_good = copy.deepcopy(left_fit)
    right_fit_last_good = copy.deepcopy(right_fit)
    left_fitx_last_good = copy.deepcopy(left_fitx)
    right_fitx_last_good=copy.deepcopy(right_fitx)
    ploty_last_good=copy.deepcopy(ploty)
    left_curverad_last_good=copy.deepcopy(left_curverad)
    right_curverad_last_good=copy.deepcopy(right_curverad)
    offcenter_last_good=copy.deepcopy(offcenter)
    
    return left_fitx,right_fitx,ploty,left_curverad,right_curverad,offcenter

def abs_sobel_thresh(gray, orient='x', thresh=(0,255)):
    thresh_min = thresh[0]
    thresh_max = thresh[1]
    # Apply x or y gradient with the OpenCV Sobel() function
    # and take the absolute value
    if orient == 'x':
        abs_sobel = np.absolute(cv2.Sobel(gray, cv2.CV_64F, 1, 0))
    if orient == 'y':
        abs_sobel = np.absolute(cv2.Sobel(gray, cv2.CV_64F, 0, 1))
    # Rescale back to 8 bit integer
    scaled_sobel = np.uint8(255*abs_sobel/np.max(abs_sobel))
    # Create a copy and apply the threshold
    binary_output = np.zeros_like(scaled_sobel)
    # Here I'm using inclusive (>=, <=) thresholds, but exclusive is ok too
    binary_output[(scaled_sobel >= thresh_min) & (scaled_sobel <= thresh_max)] = 1
    # Return the result
    return binary_output
def mag_thresh(gray, sobel_kernel=3, mag_thresh=(0, 255)):
    # Convert to grayscale
    # Take both Sobel x and y gradients
    sobelx = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=sobel_kernel)
    sobely = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=sobel_kernel)
    # Calculate the gradient magnitude
    gradmag = np.sqrt(sobelx**2 + sobely**2)
    # Rescale to 8 bit
    scale_factor = np.max(gradmag)/255 
    gradmag = (gradmag/scale_factor).astype(np.uint8) 
    # Create a binary image of ones where threshold is met, zeros otherwise
    binary_output = np.zeros_like(gradmag)
    binary_output[(gradmag >= mag_thresh[0]) & (gradmag <= mag_thresh[1])] = 1
    # Return the binary image
    return binary_output

def dir_threshold(gray, sobel_kernel=3, thresh=(-np.pi/4, np.pi/4)):
    # 2) Take the gradient in x and y separately
    sobelx = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=sobel_kernel)
    sobely = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=sobel_kernel)
    # 3) Take the absolute value of the x and y gradients
    abs_sobelx = np.absolute(sobelx)
    abs_sobely = np.absolute(sobely)
    # 4) Use np.arctan2(abs_sobely, abs_sobelx) to calculate the direction of the gradient 
    absgraddir = np.arctan2(abs_sobely,abs_sobelx)
    # 5) Create a binary mask where direction thresholds are met
    binary_output =  np.zeros_like(absgraddir)
    binary_output[(absgraddir >= thresh[0]) & (absgraddir <= thresh[1])] = 1
    # 6) Return this mask as your binary_output 
    return binary_output

def apply_thresholding(img):
    ksize=9
    hls = cv2.cvtColor(img, cv2.COLOR_RGB2HLS)
    S = hls[:,:,2]
    thresh = (130, 255)
    S_binary = np.zeros_like(S)
    S_binary[(S > thresh[0]) & (S <= thresh[1])] = 1
    
    R = img[:,:,0]
    thresh = (180, 255)
    R_binary = np.zeros_like(R)
    R_binary[(R > thresh[0]) & (R <= thresh[1])] = 1
    
   
    image = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    gradx = abs_sobel_thresh(image, orient='x',thresh=(40, 100))
    
    grady = abs_sobel_thresh(image, orient='y',thresh=(40, 100))
     
    mag_binary = mag_thresh(image, sobel_kernel=ksize, mag_thresh=(30, 100))
    dir_binary = dir_threshold(image, sobel_kernel=ksize, thresh=(1.1, 1.3))
   
    combined = np.zeros_like(dir_binary)
    #combined[S_binary==1 |(gradx == 1)  | ((mag_binary == 1) & (dir_binary == 1))] = 1
    combined[ (R_binary==1) | ((S_binary==1) & (grady == 1)) | (((gradx == 1) | (grady == 1)) & ((mag_binary == 1) & (dir_binary == 1)))] = 1
    #combined[(((gradx == 1) & (grady == 1)) | ((mag_binary == 1) | (dir_binary == 1)))] = 1
    plot_image(gradx,"gradx threshold")
    plot_image(grady,"grady threshold")
    plot_image(dir_binary,"Dir threshold")
    plot_image(mag_binary,"mag threshold")
    plot_image(S_binary,"S color threshold")
    plot_image(R_binary,"R color threshold")

    plot_image(combined,"Fully thresholded")
    return combined

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image

def corners_unwarp(img,mtx, dist,corners):
    gray=img
    # Search for corners in the grayscaled image
    img_size = (gray.shape[1], gray.shape[0])
        # For source points I'm grabbing the outer four detected corners
    src = corners
        # For destination points, I'm arbitrarily choosing some points to be
        # a nice fit for displaying our warped result 
        # again, not exact, but close enough for our purposes
    dst = np.array([[(398,633),(398,183),(933,183),(933,634)]], dtype=np.float32)
        # Given src and dst points, calculate the perspective transform matrix
    M = cv2.getPerspectiveTransform(src, dst)
    # Warp the image using OpenCV warpPerspective()
    warped = cv2.warpPerspective(gray, M, img_size)
    # Return the resulting image and matrix
    return warped, M

def find_lane_lines(img):
    imgshape = img.shape
    ysize = imgshape[0]
    xsize = imgshape[1]
    global frame_no
    dst = cv2.undistort(img, mtx, dist, None, mtx)
       
    image_int = apply_thresholding(dst)
    #return image_int
    vertices = np.array([[(50,ysize),(xsize/2 - 0.01*xsize, ysize/4+0.35*ysize), (xsize/2 + 0.02*xsize ,ysize/4+0.35*ysize), (xsize-10,ysize)]], dtype=np.int32)
    thresholded_image= region_of_interest(image_int,vertices)
    vertices = np.array([[(359,619),(532,495), (760,495), (960,619)]], dtype=np.float32)
    img_unwarped, M = corners_unwarp(thresholded_image,mtx,dist,vertices)
    plot_image(img_unwarped,"input to lane finder")
    left_fitx,right_fitx,ploty,left_curverad,right_curverad,offcenter = windowed_lane_finder(img_unwarped)
    print(left_curverad, 'm', right_curverad, 'm')
    
    warp_zero = np.zeros_like(img_unwarped).astype(np.uint8)
    color_warp = np.dstack((warp_zero, warp_zero, warp_zero))
    
    # Recast the x and y points into usable format for cv2.fillPoly()
    pts_left = np.array([np.transpose(np.vstack([left_fitx, ploty]))])
    pts_right = np.array([np.flipud(np.transpose(np.vstack([right_fitx, ploty])))])
    pts = np.hstack((pts_left, pts_right))
    
    # Draw the lane onto the warped blank image
    cv2.fillPoly(color_warp, np.int_([pts]), (0,255, 0))
    Minv = np.linalg.pinv(M)
    # Warp the blank back to original image space using inverse perspective matrix (Minv)
    newwarp = cv2.warpPerspective(color_warp, Minv, (xsize, ysize)) 
    # Combine the result with the original image
    result = cv2.addWeighted(dst, 1, newwarp, 0.3, 0)

    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(result,'Left radius of curvature  = %.2f m'%(left_curverad),(60,50), font, 1,(255,255,0),2,cv2.LINE_AA)
    cv2.putText(result,'Right radius of curvature = %.2f m'%(right_curverad),(60,80), font, 1,(255,255,0),2,cv2.LINE_AA)
    cv2.putText(result,'Off center by  %.2f m'%(offcenter),(60,110), font, 1,(255,255,0),2,cv2.LINE_AA)
    

    return result
```
